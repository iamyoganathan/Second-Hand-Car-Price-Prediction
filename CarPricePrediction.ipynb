```python
df = df.drop(df.columns[303], axis = 'columns')
df
```

```python
def normalization(feature):
    feature = feature - np.mean(feature) / np.std(feature)
    return feature

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df['Kms_driven']
```

```python
Kms_driven = np.array(df['Kms_driven'])
Kms_driven = Kms_driven.reshape(-1, 1)
Kms_driven
```

```python
df['Kms_driven'] = scaler.fit_transform(Kms_driven)
df['Kms_driven']
# Mean = 0
# Variance = 1
```

```python
df
```

```python
y = df['Price']
y
```

```python
x = df.drop(['Price'], axis = 'columns')
x
```

```python
Year = np.array(x['Year'])
Year = Year.reshape(-1, 1)
Year
```

```python
x['Year'] = scaler.fit_transform(Year)
x['Year']
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, random_state = 1)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

Model Building

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)

y_predicted = lr.predict(x_test)

print("The First Five Predicted Values are: ")
print(y_predicted[0:5])

print("The First Five Test Values are: ")
print(y_test[0:5])
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_predicted)
r2
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(lr, x, y, cv =10))
print(np.average(cross_val_score(lr, x, y, cv =10)))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(x_train, y_train)

y_predicted = lasso.predict(x_test)

print("The First Five Predicted Values are: ")
print(y_predicted[0:5])

print("The First Five Test Values are: ")
print(y_test[0:5])
```

```python
r2 = r2_score(y_test, y_predicted)
r2
```

```python
scores = cross_val_score(lr, x, y, cv =10)
print(scores)
print(np.average(scores))
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(x_train, y_train)

y_predicted = ridge.predict(x_test)

print("The First Five Predicted Values are: ")
print(y_predicted[0:5])

print("The First Five Test Values are: ")
print(y_test[0:5])
```

```python
r2 = r2_score(y_test, y_predicted)
r2
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(x_train, y_train)

y_predicted = dtr.predict(x_test)

print("The First Five Predicted Values are: ")
print(y_predicted[0:5])

print("The First Five Test Values are: ")
print(y_test[0:5])
```

```python
r2 = r2_score(y_test, y_predicted)
r2
```
